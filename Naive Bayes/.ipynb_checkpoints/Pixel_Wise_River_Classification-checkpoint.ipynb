### Calculating Mean of river and non-river class

```python
T1 = np.mean(river_sample_points, axis=0)
T2 = np.mean(non_river_sample_points, axis=0)

T_all = np.mean(all_sample_points, axis=0)
```

### Calculate Co-varience matrix for river and non-river class

```python
dev_T1 = np.zeros(river_sample_points.shape)
for i in range(river_sample_points.shape[0]) :
    for j in range(river_sample_points.shape[1]) :
        dev_T1[i][j] = river_sample_points[i][j] - T1[j]

Cov_Mat1 = (float(1) / float(river_sample_points.shape[0])) * np.dot(np.transpose(dev_T1), dev_T1)


dev_T2 = np.zeros(non_river_sample_points.shape)
for i in range(non_river_sample_points.shape[0]) :
    for j in range(non_river_sample_points.shape[1]) :
        dev_T2[i][j] = non_river_sample_points[i][j] - T2[j]
        
Cov_Mat2 = (float(1) / float(non_river_sample_points.shape[0])) * np.dot(np.transpose(dev_T2), dev_T2)


val2 = training_data - T_2
Cov_Mat2 = (float(1) / float(training_data.shape[0])) * np.dot(np.transpose(val2), val2)
```
