```python
from numpy import*
from scipy.interpolate import*
from matplotlib.pyplot import*
```

```python
%matplotlib inline
```

```python
x = array([2007, 2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017])
y = array([61.2, 58.3, 67.1, 69.2, 68.9, 83.5, 89.1, 80, 92.3, 93, 97])
tran = x.size
```

```python
plot(x, y, 'o')
show()
```

```python
p1 = polyfit(x, y, 1)
p1
```

```python
plot(x, polyval(p1, x), 'r-')
plot(x, y, 'o')
show()
```

```python
a = 0
b = 0
c = 0
d = 0
for i in range(tran):
  a += x[i]
  b += y[i]
  c += x[i]*x[i]
  d += x[i]*y[i]
```

```python
w1 = ((a*b) - (d*tran))/((a*a)-(tran*c)) 
w0 = (b - (w1*a))/tran
```

```python
p = array([w0, w1])
p
```

```python
plot(x, w0+(w1*x), 'r-')
plot(x, y, 'o')
show()
```

```python
w0 + (w1*2019)
```

```python
error = []
for i in range(tran):
  e = y[i]-polyval(p, 2019)
  error.append(e)
error  
```
